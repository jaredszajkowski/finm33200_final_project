## Explore Data

Here we provide examples of the Parquet files generated as part of the data pipeline.

```python
parquet_files = [
    "fred.parquet",
]

for pf in parquet_files:
    print(pf)
    df = pl.scan_parquet(DATA_DIR / pf)
    display(df.collect_schema())
    print(df.collect().shape)
    display(df.head().collect())
```

### task_embed

```python
bert_df = pd.read_parquet(DATA_DIR / "embeddings_bert_chunks" / "chunk_000000.parquet")
display(bert_df)
```

```python
gemma_df = pd.read_parquet(DATA_DIR / "embeddings_gemma_chunks" / "chunk_000000.parquet")
display(gemma_df)
```

```python
# find a common `rp_story_id` between the two datasets
common_rp_story_id = set(bert_df['rp_story_id']).intersection(set(gemma_df['rp_story_id']))
common_rp_story_id
```

```python
common = pd.concat([bert_df[bert_df['rp_story_id'] == "7B82075FB185D295FAB83D2467CDCB64"],
                    gemma_df[gemma_df['rp_story_id'] == "7B82075FB185D295FAB83D2467CDCB64"]])
common.reset_index(drop=True, inplace=True)

common.loc[0, "rp_story_id"] = "BERT_" + "7B82075FB185D295FAB83D2467CDCB64"
common.loc[1, "rp_story_id"] = "GEMMA_" + "7B82075FB185D295FAB83D2467CDCB64"

common.set_index("rp_story_id", inplace=True)
common
```
